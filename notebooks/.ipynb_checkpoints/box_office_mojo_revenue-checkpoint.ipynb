# TMDB: Get revenue and rating from Box office mojo

```python
df = pd.read_csv("data/raw/tmdb.csv", lineterminator = '\n')
print(df.shape)
df = df.dropna(subset = ['imdb_id'])
print(df.shape)
ids = list(df['imdb_id'])
```

```python
print(sum(df['budget'] == 0))
print(sum(df['revenue'] == 0))
```

```python

res = []
for id in tqdm(ids):

    r = requests.get(f'https://www.boxofficemojo.com/title/{id}/?ref_=bo_se_r_1')
    soup = BeautifulSoup(r.text, 'lxml')
    moneys = soup.find_all('span', attrs={'class':'a-size-medium a-text-bold'})
    
    # budget
    try:
        budget = soup.select_one('div.a-section.a-spacing-none.mojo-gutter.mojo-summary-table > div.a-section.a-spacing-none.mojo-summary-values.mojo-hidden-from-mobile > div:nth-of-type(3) > span:nth-of-type(2) > span').text
        budget = budget.replace("$","").replace(",","")
    except:
        budget = 0
    
    # revenue (domestic + international + total)
    row_revenue = []
    row_revenue.append(id)
    row_revenue.append(budget)

    for money in moneys:
        temp = money.text.strip()
        if temp == '–':
            temp = 0
        else:
            temp = temp.replace("$","").replace(",","")
        row_revenue.append(temp)
    res.append(np.array(row_revenue))

    
    # break
    # time.sleep(random.uniform(0.1, 0.6))

res
```

```python
res = np.array(res)
temp = pd.DataFrame({'imdb_id': res[:, 0], 'budget': res[:, 1], 'domestic_revenue': res[:, 2], 'internat_revenue': res[:, 3], 'total_revenue': res[:, 4]})
temp.head()
```

```python
cols = temp.columns
temp[cols] = temp[cols].replace({'0':np.nan, 0:np.nan})
df['budget'] = df['budget'].replace({0:np.nan})
df['revenue'] = df['revenue'].replace({0:np.nan})
```

```python
print(temp.isnull().sum())
print(df[['budget', 'revenue']].isnull().sum())
```
