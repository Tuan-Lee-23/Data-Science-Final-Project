```python
import requests
import pandas as pd 
import time 
import random
import json
```

# TMDB Top rated movies

```python
page = 4
current_page = 1
count = 0

url = 'https://api.themoviedb.org/3/movie/top_rated?api_key=6e7f63f528eb5934d1ef3aa836326019&language=en-US&page='
df_res = pd.DataFrame()
while current_page <= page:
    r = requests.get(url + str(current_page))
    json_res = r.text
    json_res = json.loads(json_res)
    page = json_res['total_pages']

    json_res = json_res['results']

    df = pd.DataFrame.from_dict(json_res)
    df_res = pd.concat([df_res, df])
    
    count += len(df)
    current_page += 1
    print(f"Count: {count}")
    print(f"Page: {current_page}/{page}")
```

```python
print(df_res.shape)
df_res = df_res.drop_duplicates('id').reset_index(drop = True)
print(df_res.shape)
df_res.to_csv("top_rated_tmdb.csv", index = False)
```
