# TMDB Top rated movies

```python
page = 4
current_page = 1
count = 0

url = 'https://api.themoviedb.org/3/movie/top_rated?api_key=6e7f63f528eb5934d1ef3aa836326019&language=en-US&page='
df_res = pd.DataFrame()
while current_page <= page:
    r = requests.get(url + str(current_page))
    json_res = r.text
    json_res = json.loads(json_res)
    page = json_res['total_pages']

    json_res = json_res['results']

    df = pd.DataFrame.from_dict(json_res)
    df_res = pd.concat([df_res, df])
    
    count += len(df)
    current_page += 1
    print(f"Count: {count}")
    print(f"Page: {current_page}/{page}")
```

```python
print(df_res.shape)
df_res = df_res.drop_duplicates('id').reset_index(drop = True)
print(df_res.shape)
df_res.to_csv("top_rated_tmdb.csv", index = False)
```

## Filter cast

```python
cast_df = pd.read_csv('cast_tmdb.csv', index_col = 'id')
# cast_df = cast_df.loc[cast_df['cast'].str.len() >= 1]
# cast_df = cast_df.loc[cast_df['crew'].str.len() >= 1]
cast_df.shape
```

```python
def filter_cast(val):
    # assert type(val) == 'dict', "Using str ?? Please convert your value into dict Python"
    val = eval(val)
    if len(val) < 1:
        return np.nan
        
    temp = pd.DataFrame.from_dict(val)
    cols = ['id', 'adult', 'known_for_department', 'name', 'cast_id', 'profile_path', 'credit_id', 'order']

    temp = temp.drop(cols, axis = 1)

    temp = temp.head(5).to_dict('records')
    return temp
    
def filter_crew(val):
    # assert type(val) == 'dict', "Using str ?? Please convert your value into dict Python"
    val = eval(val)
    if len(val) < 1:
        return np.nan
    temp = pd.DataFrame.from_dict(val)
    cols = ['id', 'adult', 'known_for_department', 'name', 'profile_path', 'credit_id']
    temp = temp.drop(cols, axis = 1)
    job_filtered = ['Director', 'Screenplay', 'Director of Photography']
    temp = temp.loc[temp['job'].isin(job_filtered), :]
    temp = temp.to_dict('records')
    return temp






# filter_cast(cast_df.iloc[10, 0])
# filter_crew(cast_df.iloc[30, 1])

cast_df['cast'] = cast_df['cast'].progress_apply(filter_cast)
cast_df['crew'] = cast_df['crew'].progress_apply(filter_crew)
cast_df.to_csv('cast_filtered_tmdb.csv')
```
